```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_sigmoid():
        x = np.linspace(-10, 10, 100)
        y = 1 / (1 + np.exp(-x))
        plt.plot(x, y)
        plt.xlabel('Input')
        plt.ylabel('Sigmoid Ouptut')
        plt.title('Sigmoid Activation Function')
        plt.grid(True)
        plt.show()
```

```python
def plot_tanh():
        x = np.linspace(-10, 10, 100)
        y = np.tanh(x)
        plt.plot(x, y)
        plt.title('Hyberbolic Tangent (tanh) Activation Function')
        plt.xlabel('x')
        plt.ylabel('tanh(x)')
        plt.grid(True)
        plt.show()
```

```python
def plot_relu():
        x = np.linspace(-10, 10, 100)
        y = np.maximum(0, x)
        plt.plot(x, y)
        plt.title('ReLU Activation Function')
        plt.xlabel('x')
        plt.ylabel('ReLU(x)')
        plt.grid(True)
        plt.show()
```

```python
def plot_leaky_relu():
        x = np.linspace(-10, 10, 100)
        alpha = 0.1
        y = np.where(x >= 0, x, alpha * x)
        plt.plot(x, y)
        plt.title('Leaky ReLU Activation Function')
        plt.xlabel('x')
        plt.ylabel('Leaky ReLU(x)')
        plt.grid(True)
        plt.show()
```

```python
def softmax():
        x = np.array([1, 2, 3])
        e_x = np.exp(x - np.max(x))
        y = e_x / np.sum(e_x, axis = 0)
        class_labels = ["Class A", "Class B", "Class C"]
        plt.bar(y, class_labels)
        plt.xlabel('Class')
        plt.ylabel('Probabilies')
        plt.title('Softmax Output')
        plt.show()
```

```python
while True:
        print('\nMAIN MENU:\n')
        print('1. Sigmoid')
        print('2. Hyberbolic tangent')
        print('3. Rectified Linear Unit')
        print('4. Leaky ReLU')
        print('5. Softmax')
        print('6. Exit')

        ch = int(input("Enter your choice: "))

        if ch == 1:
                plot_sigmoid()
        elif ch == 2:
                plot_tanh()
        elif ch == 3:
                plot_relu()
        elif ch == 4:
                plot_leaky_relu()
        elif ch == 5:
                softmax()
        elif ch == 6:
                break
        else:
                print("Wrong Choice!!")
```
